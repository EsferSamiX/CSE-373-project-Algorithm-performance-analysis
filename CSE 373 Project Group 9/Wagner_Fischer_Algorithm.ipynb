```python
import random
import string
import time
import matplotlib.pyplot as plt

# Generate a random string of specified length
def generate_random_string(length):
    return ''.join(random.choices(string.ascii_lowercase, k=length))

# Measure the time it takes to execute a function with two string arguments
def measure_time(func, a, b):
    start_time = time.time()
    func(a, b)
    return time.time() - start_time

# Define a placeholder for the recursive dynamic programming function
def recursive_dynamic_programming(a, b):
    m, n = len(a), len(b)
    memo = {}

    def helper(i, j):
        if (i, j) in memo:
            return memo[(i, j)]
        if i == 0:
            result = j
        elif j == 0:
            result = i
        elif a[i - 1] == b[j - 1]:
            result = helper(i - 1, j - 1)
        else:
            result = 1 + min(helper(i - 1, j), helper(i, j - 1), helper(i - 1, j - 1))
        memo[(i, j)] = result
        return result

    return helper(m, n)

# Function to run the average case scenario
def run_average_case():
    lengths = [0, 10, 20, 30]  # Include 0 length to start from (0, 0)
    constant_string = "constantstring"

    average_times = []

    for length in lengths:
        input_string = generate_random_string(length)
        average_time = measure_time(recursive_dynamic_programming, input_string, constant_string[:length])
        average_times.append(average_time)

    plt.figure(figsize=(8, 5))
    plt.plot(lengths, average_times, label='Average-case')
    plt.xlabel('Length of Input String')
    plt.ylabel('Time (seconds)')
    plt.title('Recursive Dynamic Programming - Average-case Scenario')
    plt.legend()
    plt.grid(True)
    plt.show()

run_average_case()
```

```python
import random
import string
import time
import matplotlib.pyplot as plt

def generate_random_string(length):
    return ''.join(random.choices(string.ascii_lowercase, k=length))

def measure_time(func, a, b):
    start_time = time.time()
    func(a, b)
    return time.time() - start_time
```

```python
def wagner_fischer_edit_distance(a, b):
    m, n = len(a), len(b)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if a[i - 1] == b[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = 1 + min(dp[i - 1][j], dp[i][j - 1], dp[i - 1][j - 1])
    return dp[m][n]
```

```python
def run_wagner_fischer_best_case():
    lengths = [10, 100, 1000]
    wagner_fischer_times = []

    for length in lengths:
        identical_string = generate_random_string(length)
        wagner_fischer_time = measure_time(wagner_fischer_edit_distance, identical_string, identical_string)
        wagner_fischer_times.append(wagner_fischer_time)

    plt.figure(figsize=(8, 5))
    plt.plot(lengths, wagner_fischer_times, label='Wagner-Fischer Algorithm - Best case')
    plt.xlabel('Length of Input String')
    plt.ylabel('Time (seconds)')
    plt.title('Wagner-Fischer Algorithm Edit Distance Time Complexity - Best case')
    plt.legend()
    plt.grid(True)
    plt.show()
run_wagner_fischer_best_case()
```

```python
def run_wagner_fischer_average_case():
    lengths = [10, 100, 1000]
    wagner_fischer_times = []

    for length in lengths:
        random_string1 = generate_random_string(length)
        random_string2 = generate_random_string(length)
        wagner_fischer_time = measure_time(wagner_fischer_edit_distance, random_string1, random_string2)
        wagner_fischer_times.append(wagner_fischer_time)

    plt.figure(figsize=(8, 5))
    plt.plot(lengths, wagner_fischer_times, label='Wagner-Fischer Algorithm - Average case')
    plt.xlabel('Length of Input String')
    plt.ylabel('Time (seconds)')
    plt.title('Wagner-Fischer Algorithm Edit Distance Time Complexity - Average case')
    plt.legend()
    plt.grid(True)
    plt.show()
run_wagner_fischer_average_case()
```

```python
def run_wagner_fischer_worst_case():
    lengths = [10, 100, 1000]
    wagner_fischer_times = []

    for length in lengths:
        input_string = generate_random_string(length)
        different_string = ''.join(random.choices(string.ascii_lowercase, k=length))
        wagner_fischer_time = measure_time(wagner_fischer_edit_distance, input_string, different_string)
        wagner_fischer_times.append(wagner_fischer_time)

    plt.figure(figsize=(8, 5))
    plt.plot(lengths, wagner_fischer_times, label='Wagner-Fischer Algorithm - Worst case')
    plt.xlabel('Length of Input String')
    plt.ylabel('Time (seconds)')
    plt.title('Wagner-Fischer Algorithm Edit Distance Time Complexity - Worst case')
    plt.legend()
    plt.grid(True)
    plt.show()
run_wagner_fischer_worst_case()
```
