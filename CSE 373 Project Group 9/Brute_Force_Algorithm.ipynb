```python
import random
import string
import time
import matplotlib.pyplot as plt

def generate_random_string(length):
    return ''.join(random.choices(string.ascii_lowercase, k=length))

def measure_time(func, a, b):
    start_time = time.time()
    func(a, b)
    return time.time() - start_time
```

```python
def brute_force_edit_distance(a, b, m, n):
    if m == 0:
        return n
    if n == 0:
        return m
    if a[m - 1] == b[n - 1]:
        return brute_force_edit_distance(a, b, m - 1, n - 1)
    return 1 + min(
        brute_force_edit_distance(a, b, m, n - 1),
        brute_force_edit_distance(a, b, m - 1, n),
        brute_force_edit_distance(a, b, m - 1, n - 1)
    )
```

```python
#this
def run_brute_force():
    lengths = [2, 4, 6]
    constant_string = "aaaaaa"  # Modify constant string to an identical string

    recursive_times = []

    for length in lengths:
        random_string = constant_string  # Set random string identical to constant string
        recursive_time = measure_time(lambda a, b: brute_force_edit_distance(a, b, len(a), len(b)), random_string, constant_string)
        recursive_times.append(recursive_time)

    plt.figure(figsize=(8, 5))  # Adjust the figsize here
    plt.plot(lengths, recursive_times, label='Recursive Brute Force-Best case')
    plt.xlabel('Length of Input String')
    plt.ylabel('Time (seconds)')
    plt.title('Recursive Brute Force Edit Distance Time Complexity')
    plt.legend()
    plt.grid(True)
    plt.show()

run_brute_force()
```

```python
def run_brute_force_average_case():
    lengths = [2, 4, 6]
    constant_string = "constantstring"

    recursive_times = []

    for length in lengths:
        random_string = generate_random_string(length)
        recursive_time = measure_time(lambda a, b: brute_force_edit_distance(a, b, len(a), len(b)), random_string, constant_string)
        recursive_times.append(recursive_time)

    plt.figure(figsize=(8, 5))
    plt.plot(lengths, recursive_times, label='Brute Force - Average case')
    plt.xlabel('Length of Input String')
    plt.ylabel('Time (seconds)')
    plt.title('Brute Force Edit Distance Time Complexity - Average case')
    plt.legend()
    plt.grid(True)
    plt.show()

run_brute_force_average_case()
```

```python
def run_brute_force_worst_case():
    lengths = [4,8,10]  # Use longer input string lengths
    constant_string = "constantstring"

    recursive_times = []

    for length in lengths:
        random_string = generate_random_string(length)
        recursive_time = measure_time(lambda a, b: brute_force_edit_distance(a, b, len(a), len(b)), random_string, constant_string)
        recursive_times.append(recursive_time)

    plt.figure(figsize=(8, 5))
    plt.plot(lengths, recursive_times, label='Brute Force - Worst case')
    plt.xlabel('Length of Input String')
    plt.ylabel('Time (seconds)')
    plt.title('Brute Force Edit Distance Time Complexity - Worst case')
    plt.legend()
    plt.grid(True)
    plt.show()

run_brute_force_worst_case()

```
